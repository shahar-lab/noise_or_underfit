```python
from imports import*
from rnn import *

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
from tqdm import tqdm

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
num_of_blocks = 5
num_of_agents = 100
array = np.arange(num_of_blocks)
cv = [np.roll(array,i) for i in range(num_of_blocks)]
cv = np.array(cv)

n_cells = [5]
epochs = [100,200,400,600,1000]
```

```python
models = ['mf','mb','habit','wsls','kdh']

for model in models:
    print(f'** Model {model} **')
    
    data_results = {
                    'cell':[],
                    'epoch':[],
                    'agent': [],
                    'train_nlp_rnn' : [],
                    'test_acc_rnn':[],
                    'test_nlp_rnn': []
                   }
    
    for cell in n_cells:
        print(f'** cell {cell} **')
        
        test_loss = []
        for epoch in tqdm(epochs):
            print(f'** epoch {epoch} **')
            
            tt_nlp = 0
            for agent in range(num_of_agents):
                print(f'*agent {agent}*')   
                df_s = []
                path = f'../data/{model}/{model}_agent_{agent}'
                
                for s in range(5):
                    df_s.append(pd.read_csv(f'{path}_sim_{s}.csv'))
                    
                nlp = 0
                for n,t in enumerate(cv):
                    train_arr = t[0:-1]
                    test_arr = t[-1:]

                    # split train and test data
                    train_data = [df_s[sim] for sim in train_arr]
                    train_data = pd.concat(train_data) 
                    train_data.reset_index(inplace=True)
                    n_train = len(train_data)
                    train_data = behavior_dataset(train_data,length=n_train,if_state=True,if_state_one_back=True,if_trial=False)

                    test_data = [df_s[sim] for sim in test_arr]
                    test_data = pd.concat(test_data) 
                    test_data.reset_index(inplace=True)
                    n_test = len(test_data)
                    test_data = behavior_dataset(test_data,length=n_test,if_state=True,if_state_one_back=True,if_trial=False)

                    # train
                    train_loader = DataLoader(train_data,shuffle=False,batch_size=n_train)

                    rnn = GRU_NN(input_size=5, hidden_size=cell, num_layers=1, output_size=2)

                    rnn,train_loss = train_model(rnn,
                                                 train_loader,
                                                 device,
                                                 lr=0.001,
                                                 epochs=epoch,
                                                 input_size=5,
                                                 output_size=2)

                    cur_loss, acc =  eval_network(rnn,test_data,input_size=5)
                    nlp += cur_loss

                    data_results['cell'].append(cell)
                    data_results['epoch'].append(epoch)
                    data_results['agent'].append(agent)
                    data_results['train_nlp_rnn'].append(train_loss)
                    data_results['test_nlp_rnn'].append(cur_loss)
                    data_results['test_acc_rnn'].append(acc)
                    
                tt_nlp += nlp    
            test_loss.append(tt_nlp/(num_of_agents*num_of_blocks))
            print(test_loss)
    df = pd.DataFrame(data_results)
    df.to_csv(f'../results/{model}/rnn_fit_{model}.csv')
```
