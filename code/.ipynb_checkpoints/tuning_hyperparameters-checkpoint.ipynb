```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from tqdm import tqdm

import tensorflow as tf
from plot_stats import *
from utils import *
from rnn import *

print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
train_acc = []
test_acc = []
test_loss = []
num_of_trials = 7*50 # 

cells = 5
batch = 50
epoch = [100,400,600,1000]

#config num of blocks for each agent 
num_of_simulation = 8
# for cross valdation 
array = np.arange(num_of_simulation)
cv = [np.roll(array,i) for i in range(num_of_simulation)]
cv = np.array(cv)

for cell in tqdm(cells):
    print(f'*** cell {cell} ***')
    
    tt_acc = 0
    tt_acc_t = 0
    tt_loss = 0
    
    for a in range(50):
        print(f'agent {a}')
        
        for t in cv:
            train_arr = t[0:-1]
            test_arr = t[-1:]

            path = f'../data/hybrid/sim_data/hybrid_agent_{a}'

            model, sim_model, hist = train_a_model(
                                                      path,
                                                      train_arr,
                                                      test_arr,
                                                      num_of_trials=350,
                                                      n_cells=cell,
                                                      lr=0.001,
                                                      epochs=epoch,
                                                      batch_size=batch,
                                                      cp_callback=True
                                            )

            tt_acc += hist.history['accuracy'][-1]
            tt_acc_t += hist.history['val_accuracy'][-1]
            tt_loss += hist.history['val_binary_crossentropy'][-1]
    
    train_acc.append(tt_acc/100)  
    test_acc.append(tt_acc_t/100)
    test_loss.append(tt_loss/100)
    print(test_loss)
```
